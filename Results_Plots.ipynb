```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import os
import matplotlib.ticker as mtick
```

```python
results_folder = '/home/lawrence/Personal/Masters/COMP0087_ Natural_Language_Processing/Project/Data/Results/'
k_values = [2, 3, 4, 5, 6, 7, 8, 9, 10, 15, 20, 25, 30, 35, 40, 45, 50, 60, 70, 80, 100]
```

# Regressions

```python
plt.figure(figsize=[27, 22.5])
fs = 20
ts = 35

plt.subplot(3, 3, 1)
p_file = open(results_folder + 'KMeans_R2_Returns', 'rb')
r2_returns_results_km = pickle.load(p_file)
p_file.close()

mu = np.mean(np.array(r2_returns_results_km), axis=1)
ss = np.std(np.array(r2_returns_results_km), axis=1) / np.sqrt(200)
plt.plot(k_values, mu, 'b')
plt.legend(['K-Means'], fontsize=fs)
plt.plot(k_values, mu + ss, 'b:')
plt.plot(k_values, mu - ss, 'b:')
plt.title('Return Regression $R^2$', fontsize=ts)
plt.ylabel('$R^2$', fontsize=fs)
plt.ylim([-.02, .1])
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs);
ax = plt.gca()
ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0, decimals=0));

plt.subplot(3, 3, 2)
p_file = open(results_folder + 'KMeans_R2_Risk', 'rb')
r2_stdev_results_km = pickle.load(p_file)
p_file.close()

mu = np.mean(np.array(r2_stdev_results_km), axis=1)
ss = np.std(np.array(r2_stdev_results_km), axis=1) / np.sqrt(200)
plt.plot(k_values, mu, 'b')
plt.legend(['K-Means'], fontsize=fs)
plt.plot(k_values, mu + ss, 'b:')
plt.plot(k_values, mu - ss, 'b:')
plt.title('Risk Regression $R^2$', fontsize=ts)
plt.ylabel('$R^2$', fontsize=fs)
plt.ylim([0, 0.30])

plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs);
ax = plt.gca()
ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0, decimals=0));

plt.subplot(3, 3, 3)
p_file = open(results_folder + 'KMeans_Coherence', 'rb')
tc_values_km = pickle.load(p_file)
p_file.close()
mu = np.mean(np.array(tc_values_km), axis=1)
ss = np.std(np.array(tc_values_km), axis=1) / np.sqrt(200)
plt.plot(k_values, mu, 'b')
plt.plot(k_values, mu + ss, 'b:')
plt.legend(['K Means'], fontsize=fs)
plt.plot(k_values, mu - ss, 'b:')
plt.ylabel('Coherence', fontsize=fs)
plt.title('Coherence', fontsize=ts)
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs)

plt.subplot(3, 3, 4)
p_file = open(results_folder + 'LDA_R2_Returns', 'rb')
r2_stdev_results_km = pickle.load(p_file)
p_file.close()

mu = np.median(np.array(r2_stdev_results_km), axis=1)
ss = np.median(np.abs(np.array(r2_stdev_results_km) - mu.reshape((-1, 1))), axis=1) / np.sqrt(200)
plt.plot(k_values, mu, 'b')
plt.legend(['LDA'], fontsize=fs)
plt.plot(k_values, mu + ss, 'b:')
plt.plot(k_values, mu - ss, 'b:')
plt.ylabel('$R^2$', fontsize=fs)
plt.xticks(fontsize=fs)
plt.xlim([0, 100])
plt.yticks(fontsize=fs);
plt.ylim([-.02, .1])
ax = plt.gca()
ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0, decimals=0));

plt.subplot(3, 3, 5)
p_file = open(results_folder + 'LDA_R2_Risk', 'rb')
r2_stdev_results_km = pickle.load(p_file)
p_file.close()

mu = np.median(np.array(r2_stdev_results_km), axis=1)
ss = np.median(np.abs(np.array(r2_stdev_results_km) - mu.reshape((-1, 1))), axis=1) / np.sqrt(20)
plt.plot(k_values, mu, 'b')
plt.legend(['LDA'], fontsize=fs)
plt.plot(k_values, mu + ss, 'b:')
plt.plot(k_values, mu - ss, 'b:')
plt.ylabel('$R^2$', fontsize=fs)
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs);
plt.ylim([0, 0.30])
ax = plt.gca()
ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0, decimals=0));

plt.subplot(3, 3, 6)
p_file = open(results_folder + 'LDA_Coherence', 'rb')
tc_values_km = pickle.load(p_file)
p_file.close()
mu = np.mean(np.array(tc_values_km), axis=1)
ss = np.std(np.array(tc_values_km), axis=1) / np.sqrt(200)
plt.plot(k_values, mu, 'b')
plt.plot(k_values, mu + ss, 'b:')
plt.legend(['LDA'], fontsize=fs)
plt.plot(k_values, mu - ss, 'b:')
plt.ylabel('Coherence', fontsize=fs)
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs)

plt.subplot(3, 3, 7)
p_file = open(results_folder + 'NVDM_R2_Returns_100', 'rb')
r2_stdev_results_km = pickle.load(p_file)
p_file.close()

mu = np.mean(np.array(r2_stdev_results_km), axis=1)
ss = np.std(np.array(r2_stdev_results_km), axis=1) / np.sqrt(200)
plt.plot(k_values, mu, 'b')
plt.legend(['NVDM'], fontsize=fs)
plt.plot(k_values, mu + ss, 'b:')
plt.plot(k_values, mu - ss, 'b:')
plt.xlabel('K', fontsize=fs)
plt.ylabel('$R^2$', fontsize=fs)
plt.xticks(fontsize=fs)
plt.xlim([0, 100])
plt.ylim([-.02, .1])
plt.yticks(fontsize=fs);
ax = plt.gca()
ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0, decimals=0));

plt.subplot(3, 3, 8)
p_file = open(results_folder + 'NVDM_R2_Risk_100', 'rb')
r2_stdev_results_km = pickle.load(p_file)
p_file.close()

mu = np.mean(np.array(r2_stdev_results_km), axis=1)
ss = np.std(np.array(r2_stdev_results_km), axis=1) / np.sqrt(200)
plt.plot(k_values, mu, 'b')
plt.legend(['NVDM'], fontsize=fs)
plt.plot(k_values, mu + ss, 'b:')
plt.plot(k_values, mu - ss, 'b:')
plt.xlabel('K', fontsize=fs)
plt.ylabel('$R^2$', fontsize=fs)
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs);
plt.ylim([0, 0.30])
ax = plt.gca()
ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0, decimals=0));

plt.subplot(3, 3, 9)
p_file = open(results_folder + 'NVDM_Coherence_100', 'rb')
tc_values_km = pickle.load(p_file)
p_file.close()
mu = np.mean(np.array(tc_values_km), axis=1)
ss = np.std(np.array(tc_values_km), axis=1) / np.sqrt(200)
plt.plot(k_values, mu, 'b')
plt.plot(k_values, mu + ss, 'b:')
plt.legend(['NVDM'], fontsize=fs)
plt.plot(k_values, mu - ss, 'b:')
plt.xlabel('K', fontsize=fs)
plt.ylabel('Coherence', fontsize=fs)
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs)

plt.savefig('/home/lawrence/Personal/Masters/COMP0087_ Natural_Language_Processing/Project/Pictures/Results_All.png')
```
